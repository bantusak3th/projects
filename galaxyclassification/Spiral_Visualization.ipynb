```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import keras
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D
from keras.models import Sequential
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
spiral = cv2.imread("/PGC0000282.png")
```

```python
plt.imshow(spiral)
```

```python
spiral.shape
```

```python
num_classes=3
```

```python
model1 = Sequential()
model1.add(Conv2D(3,5,5, activation='tanh', input_shape=spiral.shape))
```

```python
def visualize_spi(model, spiral):
    fig = plt.gcf()
    spi_batch = np.expand_dims(spiral,axis=0)
    conv_spi = model.predict(spi_batch)
    conv_spi = np.squeeze(conv_spi, axis=0)
    print (conv_spi.shape)
    plt.imshow(conv_spi)
    fig.savefig('3_filter_1st_layer_visuals_spi.png')
```

```python
visualize_spi(model1, spiral)
```

```python
model2 = Sequential()
model2.add(Conv2D(1,5,5, activation='tanh', input_shape=spiral.shape))
model2.add(MaxPooling2D(pool_size=(5,5)))
```

```python
def nice_galaxy_printer(model, spiral):
    fig = plt.gcf()
    spi_batch = np.expand_dims(spiral,axis=0)
    conv_spi2 = model.predict(spi_batch)

    conv_spi2 = np.squeeze(conv_spi2, axis=0)
    print (conv_spi2.shape)
    conv_spi2 = conv_spi2.reshape(conv_spi2.shape[:2])

    print (conv_spi2.shape)
    plt.imshow(conv_spi2)
    fig.savefig('1_filter_1st_layer_visuals_with_pooling_spi.png')
```

```python
nice_galaxy_printer(model2, spiral)
```

```python
model3 = Sequential()
model3.add(Conv2D(3,5,5, activation='tanh', input_shape=spiral.shape))
model3.add(MaxPooling2D(pool_size=(5,5)))
model3.add(Conv2D(3,5,5, activation='tanh', input_shape=spiral.shape))
model3.add(MaxPooling2D(pool_size=(5,5)))

visualize_spi(model3, spiral)
```
